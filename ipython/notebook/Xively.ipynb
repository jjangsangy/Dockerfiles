## Parameters

Xively REST API currently supports 5 parameter argumnents for reading logger info

- stream
- duration
- interval
- limit
- filetype

## Backend Implementation

```python
# -*- coding: utf-8 -*-
from __future__ import print_function, unicode_literals

import dateutil.parser
import json
import requests
import sys

from IPython.html.widgets import interact, interactive, fixed
from IPython.utils.traitlets import link
from IPython.html import widgets
from IPython.display import display

# Don't generate *.pyc/__pycache__ for single file programs
sys.dont_write_bytecode = True

API_KEY = "2jAqS1oCOKTQMrn84UIgMp7kpHoYhZFSV0acxsb3fntyANwX"
FEED    = "1323604224"

class Therm(object):
    """
    Author: Sang Han
    Calient Technologies 2014

    Only supports a minimal amount of features necessary
    for collecting and formatting data and HTTP/1.1 GET Requests.
    """
    def __init__(self, stream=None, duration=None, interval=None, limit=100):
        self.stream = stream
        self.duration = duration
        self.interval = interval
        self.limit = limit

    def connect(self):
        """Composes a request to Xively API endpoint and returns a response object"""
        base = 'https://api.xively.com/v2/feeds'
        url  = '{base}/{feed}.{ext}'.format(
                base=base, feed=FEED, ext=self.filetype.lower()
        )
        headers = {
            "X-ApiKey":  API_KEY,
                "Host": "api.xively.com",
        }
        params = {
            "datastreams": self.stream,
               "duration": self.duration,
               "interval": self.interval,
                  "limit": self.limit
        }
        return requests.get(url, headers=headers, params=params)

    def to_csv(self):
        """Wrapper around data retrieval for csv files by overloading self.filetype"""
        self.filetype = 'csv'
        return self.connect()

    def to_json(self):
        """Wrapper around data retrieval for json files by overloading self.filetype"""
        self.filetype = 'json'
        return self.connect()

    def get_json(self):
        """Returns the JSON formatted object"""
        self.request = self.to_json()
        if self.validate_status_code():
            return json.dumps(self.request.json(), indent=4)

    def print_json(self):
        """Pretty Print JSON object for terminal output"""
        print(self.get_json())

    def to_xml(self):
        """Wrapper around data retrieval for xml files by overloading self.filetype"""
        self.filetype = 'xml'
        return self.connect()

    def validate_status_code(self):
        """
        Validate status code returned from self.connect() method. Assumes that self.request
        has already been overloaded beforehand and will raise NameError exception. otherwise.
        """
        if self.request.ok and self.request.text == "":
            print("No data exists at the range specified")
            return False
        elif self.request.status_code == 403:
            print("""
                You have specified a range that exceeded the maximum allowable values
                Please modify the `--duration` and --interval
            """)
            return False
        elif self.request.status_code != requests.codes.ok:
            return self.request.raise_for_status()
        elif self.request.ok:
            return True
        else:
            print("Program failed to execute")
            return False
        
    def print_csv(self):
        """Pretty print for terminal output"""
        self.request = self.to_csv()
        if self.validate_status_code():
            points = (data.split(',') for data in self.request.text.split('\n'))
            # Header
            print('{date:10} {time:9} {device:9} {temperature:10}'.format(
                       date = 'Date',
                       time = 'Time',
                temperature = 'Temp F',
                     device = 'Device',
            ))
            # Values
            for point in points:
                print(dateutil.parser.parse(point[1].replace('T', ' ')).ctime()[0:-4],
                    '{device:9} {temperature:0= 6.2f}'.format(
                             device = point[0],
                        temperature = float(point[2])
            ))
```

## Interval Handler

```python
def interval_conversion(duration, duration_unit):
    """Takes in duration and units of time and returns the minimum allowable interval"""
    interval = {
        'second': 1,
        'minute': 60,
        'hour': 60 * 60,
        'day': 60 * 60 * 24,
    }
    return (30 * duration * interval[duration_unit.lower().rstrip('s')]) / 43200

def submit_values(vals):
    therm = Therm(
        duration = '{duration_param}{duration_specifier}'.format(
            duration_param = duration_param.value,
            duration_specifier = duration_specifier.value.lower()
        ),
        stream = stream_param.value,
        interval = interval_conversion(duration_param.value, duration_specifier.value.lower())
    )
    if filetype_param.value.lower() == 'csv':
        return therm.print_csv()
    elif filetype_param.value.lower() == 'json':
        return therm.print_json()
    else:
        return "Filetype not correctly specified"
```
